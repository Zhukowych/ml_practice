```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()


from sklearn.datasets import make_moons
```

```python
X, y = make_moons(150, noise=0.01, random_state=42)
X = pd.DataFrame(X)
y = pd.Series(y)

sns.scatterplot(data=X, x=0, y=1, hue=y)
```
