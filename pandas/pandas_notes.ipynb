## Maps

with `map` we can transform a single `Series` object

```python
review_points_mean = wine_reviews.points.mean()
wine_reviews.points.map(lambda p: p - review_points_mean)
```

with `apply` we can change whole data frame. If we pass `index` into axis instead of `columns` we must pass a function that must change column instead of row

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

wine_reviews.apply(remean_points, axis='columns')
```

with `transform` we can pass function that will be applied to each cell in data frame


## Grouping

```python
wine_reviews.groupby('points').points.count()
```

when we make `groupby` it created a bunch of inner-dataframes which grouped by same values of arguments 

```python
wine_reviews.groupby('winery').apply(lambda df: df.title.iloc[0], include_groups=False)
```

we can also aggregate by custom functions

```python
wine_reviews.groupby(['country']).price.agg([len, 'min', 'max'])
```

### MultiIndex

Grouping produces data frames that have MultiIndex, so to get to the particular element, you must pass two indices, but we can reduce this path

```python
countries_reviewed = wine_reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
countries_reviewed.reset_index()
```

## Sorting

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

By default `sort_values` sorts in ascending order, to do the reverse we can pass an additional argument

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

We can also sort by index, or sort by multiple fields

```python
countries_reviewed.sort_index()
```

```python
countries_reviewed.sort_values(by=['country', 'len'])
```

```python
wine_reviews.groupby(['country', 'variety']).size().sort_values(ascending=False)
```

## Combining

when we have several datasets, but with same structure (i.e. same columns) we would want to stack them upon each other


```python
canadian_youtube = pd.read_csv('CAvideos.csv')
german_youtube = pd.read_csv('DEvideos.csv')

pd.concat([canadian_youtube, german_youtube])
```

If you want to combine two datasets horizontaly that have an index in common, you can do that with `join`

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = german_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix="_CAN", rsuffix="_DE")
```
